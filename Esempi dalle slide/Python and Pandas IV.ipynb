## Plotting and visualization

```python
# Import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Plotting function in pandas

```python
s = pd.Series(np.random.randn(10).cumsum(), index = np.arange(0, 100, 10))
print("Series: {}".format(s))
# La funzione 'cumsum' somma lungo l'asse specificato
s.plot()
```

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0), columns = ['A', 'B', 'C', 'D'], 
                  index = np.arange(0, 100, 10))
print('DataFrame: {}'.format(df))
df.plot()
```
